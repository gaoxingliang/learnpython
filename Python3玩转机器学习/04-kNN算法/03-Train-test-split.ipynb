```python
test_indexes = shuffle_indexs[:test_size]
```

```python
train_indexes = shuffle_indexs[test_size:]
```

```python
X_train = X[train_indexes]
y_train = y[train_indexes]
X_test = X[test_indexes]
y_test = y[test_indexes]
```
