# 梯度法

### 梯度上升法用于最大化效用函数
### 梯度下降法用于最小化损失函数

#### eta 是学习率 $\eta$

![gd1](img/gd1.png)

### 并不是所有函数都有唯一极值点
#### 多次运行随机选择初始点

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plot_x = np.linspace(-1, 6, 141)
```

```python
plot_x
```

```python
plot_y = (plot_x - 2.5) ** 2 - 1
```

```python
plt.plot(plot_x, plot_y)
```

```python
# 定义导数函数
def dJ(theta):
    return 2 * (theta - 2.5)
```

```python
# 定义损失函数
def J(theta):
    try:
        return (theta - 2.5) ** 2 - 1
    except:
        return float('inf')
```

```python
# 定义学习率
eta = 0.1

epsilon = 1e-8

theta = 0.0
while True:
    gradient = dJ(theta)
    last_theta = theta
    # 向反方向前进
    theta = theta - eta * gradient
    if (abs(J(theta) - J(last_theta)) < epsilon):
        break

print(J(theta))
print(theta)
    
```

```python
eta = 0.1

epsilon = 1e-8

theta = 0.0
theta_history = [theta]
while True:
    gradient = dJ(theta)
    last_theta = theta
    # 向反方向前进
    theta = theta - eta * gradient
    theta_history.append(theta)
    if (abs(J(theta) - J(last_theta)) < epsilon):
        break
        
plt.plot(plot_x, J(plot_x))
plt.plot(np.array(theta_history), J(np.array(theta_history)), color="red", marker="+")
```

```python
def gradient_descent(initial_theta, eta, n_iters = 1e2, epsilon=1e-8):
    runs = 0
    theta = initial_theta
    i_iter = 0
    while i_iter < n_iters:
        gradient = dJ(theta)
        i_iter += 1
        
        last_theta = theta
        # 向反方向前进
        theta = theta - eta * gradient
        theta_history.append(theta)
        if (abs(J(theta) - J(last_theta)) < epsilon):
            break
        
    #print("gd finished", len(theta_history))
    #return theta_history
        
def plot_theta_history():
    plt.plot(plot_x, J(plot_x))
    plt.plot(np.array(theta_history), J(np.array(theta_history)), color="red", marker="+")
    plt.show()
```

```python
eta = 0.01
theta_history = []
gradient_descent(0., eta)
plot_theta_history()
```
