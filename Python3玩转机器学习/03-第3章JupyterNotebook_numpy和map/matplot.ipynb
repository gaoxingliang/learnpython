# matplot lib 

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0, 10, 100)
```

```python
y = np.sin(x)
```

```python
plt.plot(x, y)
```

```python
plt.show()
```

```python
# 多条曲线
cosy = np.cos(x)
```

```python
cosy.shape
```

```python
siny = y.copy()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy)
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red")
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red")
plt.xlim(-5, 15)
plt.show()
```
